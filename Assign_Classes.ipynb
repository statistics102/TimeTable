```python
import pandas as pd
import os
```

```python
Hour = list(range(8,16,2))
Days = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu']
```

```python
pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv(r"CS_Plan.csv")
```

```python
courses = pd.DataFrame()
courses['CourseCode'] = df['Course_Code']
courses['SessionsCount']=df['Actual_Weight'].apply(lambda x:6 if x==15 else 3 if x==0 else x)
courses['LabsCount']=df['Lab_Weight']
courses['RoomsCount']=df['#Normal_Sec']
courses['Time'] = courses.apply(lambda x:pd.DataFrame(index=Days,columns=Hour), axis=1)
courses
```

```python
locations = pd.read_csv(r"Location.csv")
```

```python
locations['Time'] = locations.apply(lambda x:pd.DataFrame(index=Days,columns=Hour), axis=1)
locations
```

```python
def dayHasEnoughSpaces(dayName, RoomsCount, classType):
    count = 0
    for classroom in locations[locations.Type == classType].iterrows():
        for time, ClassCode in classroom[1].Time.loc[dayName].iteritems():
            if ClassCode != ClassCode: 
                count+=1;
                if count >= RoomsCount: return True
    return False
```

```python
def registerCourseInDay(dayName, RoomsCount, classType, CourseCode, courseIndex):
    count = 0
    for classroom in locations[locations.Type == classType].iterrows():
        for time, ClassCode in classroom[1].Time.loc[dayName].iteritems():
            if ClassCode != ClassCode: 
                count+=1
                classroom[1].Time.loc[dayName,time] = CourseCode
                courses.loc[courseIndex,'Time'].loc[dayName,time] = classroom[1].ID
                if count == RoomsCount:
                    return
```

```python
def assignClasses():
    for course in courses.iterrows():
        dayIndex = -1
        for sessionNumber in range(course[1].SessionsCount):
            dayIndex += 1
            if dayIndex == 5: dayIndex = 0
            while not dayHasEnoughSpaces(Days[dayIndex],course[1].RoomsCount, 'Classroom'):
                dayIndex+=1
                if dayIndex == 5:print('Running Out Of Days.');return
            registerCourseInDay(Days[dayIndex], course[1].RoomsCount, 'Classroom', course[1].CourseCode, course[0])
        
        dayIndex = -1
        for labNumber in range(course[1].LabsCount):
            dayIndex += 1
            if dayIndex == 5: dayIndex = 0
            while not dayHasEnoughSpaces(Days[dayIndex],course[1].RoomsCount, 'Lab'):
                dayIndex+=1
                if dayIndex == 5:print('Running Out Of Days.');return
            registerCourseInDay(Days[dayIndex], course[1].RoomsCount, 'Lab', course[1].CourseCode, course[0])
assignClasses()
```
